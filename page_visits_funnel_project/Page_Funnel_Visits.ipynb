```python
import pandas as pd
```

Import all the files

```python
visits = pd.read_csv('visits.csv',
                     parse_dates=[1])
cart = pd.read_csv('cart.csv',
                   parse_dates=[1])
                   
checkout = pd.read_csv('checkout.csv',
                       parse_dates=[1])
purchase = pd.read_csv('purchase.csv',
                       parse_dates=[1])
```

Step 1: Inspect the DataFrames using `print` and `head`

```python
print(purchase)
# print(cart.head(5))
# print(checkout.head(5))
#print(purchase.head(5))
```

Step 2: Left merging visits and cart

```python
visits_cart = visits.merge(cart, how="left")
print(visits_cart[visits_cart.cart_time.notnull()]) # visits y cart
```

Step 3: How long is `visits_cart`?

```python
print(visits_cart.head())
```

Step 4: How many timestamps are null for `cart_time`?

```python
null_cart_times = visits_cart[visits_cart.cart_time.isnull()]
print(len(null_cart_times))
```

Step 5: What percentage only visited?

```python
print((len(visits) - len(null_cart_times)) * 100 / len(visits))
```

Step 6: What percentage placed a t-shirt in their cart but did not checkout?

```python
cart_checkout = cart.merge(checkout, how='left')
#print(cart_checkout)
# no hay null values en cart_time
print(cart_checkout[cart_checkout.checkout_time.notnull()])
did_checkout = cart_checkout[cart_checkout.checkout_time.notnull()]
# did_cart_and_checkout = cart_checkout[cart_checkout.checkout_time.notnull()]
#print(did_cart_and_checkout)
```

Step 7: Merge it all together

```python
did_purchase = did_checkout.merge(purchase)

# csv
# 1. visits --> 2. cart --> 3. checkout --> 4. purchase
visits_cart = visits.merge(cart, how="left") 

cart_checkout = checkout.merge(visits_cart, how="left")

checkout_purchase = cart_checkout.merge(purchase, how="left")
not_purchase = checkout_purchase[checkout_purchase.purchase_time.isnull()]

print(len(not_purchase))
```
